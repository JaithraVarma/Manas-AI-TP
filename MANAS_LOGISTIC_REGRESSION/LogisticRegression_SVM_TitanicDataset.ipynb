```python
min_max_scaler = preprocessing.MinMaxScaler() # feature scaling using min max scaler
x_scaled = min_max_scaler.fit_transform(X)
X = pd.DataFrame(x_scaled)
X_train, X_test_trainset,Y_train,Y_test_trainset= train_test_split(X, Y,test_size=0.30,random_state=1)
```

```python
X_train1 = np.array(X_train)
```

```python
X_train1.shape
```

```python
X_test_trainset = np.array(X_test_trainset)
Y_test_trainset = np.array(Y_test_trainset)
```

```python
Y_train1 = np.array(Y_train).T
```

```python
Y_train1
```

```python
def LogisticModel(X,Y,learning_rate,iteration):
    m,n = X.shape 
    theta = np.random.rand(n)
    for i in range(iterations):
        hypofunc = 1 / ( 1 + np.exp(-(np.dot(X,theta))))
        tmp = (hypofunc - Y.T)        
        #tmp = np.reshape( tmp, m )
        dtheta = (np.dot( X.T, tmp ))/m         
        theta = theta - learning_rate * dtheta
        learning_rate = learning_rate - 0.00001
        
    return theta
        
def predict(X,theta) :
        Z = 1 / ( 1 + np.exp( - (X.dot(theta)) ) )        
        Y = np.where( Z >= 0.5, 1, 0 )        
        return Y
def eval(X_test,theta,Y_test):
    Y_pred = predict( X_test,theta )    
    correctly_classified = 0
    for count in range( np.size( Y_pred ) ) :  
        
        if Y_test[count] == Y_pred[count] :            
            correctly_classified = correctly_classified + 1              
        count = count + 1
          
    print( "Accuracy on test set by our model       :  ", (correctly_classified / count) * 100 )
```

```python
iterations = 1000
learning_rate = 0.01
theta = LogisticModel(X_train1, Y_train1, learning_rate = learning_rate, iteration = iterations)
theta
```

```python
Y_pred = predict(X_test_trainset,theta)
Y_pred
```

```python
eval(X_test_trainset,theta,Y_test_trainset)
```
